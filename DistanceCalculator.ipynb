```python
R= 6371 # radius of earth in the form of kilometer 
```

```python
# define a function that convert degree into radians 
def deg2rad(deg):
    return deg*np.pi/180
```

```python
# define a function that calculate distance 
def dist (lat1,lon1, lat2, lon2):
    d_lat=deg2rad(lat2-lat1)
    d_long=deg2rad(lon2-lon1)
    a= np.sin(d_lat/2) * np.sin(d_lat/2) + np.cos(deg2rad(lat1)) * np.cos(deg2rad(lat2)) * np.sin(d_long/2) * np.sin(d_long/2)
    c= 2 * np.arctan2(np.sqrt(a),np.sqrt(1-a))
    d= R * c
    return d
```

```python
df.columns
```

```python
for i in range(len(df)):
    df.loc[i,'distance']=dist(df.loc[i,'Restaurant_latitude'],
                               df.loc[i,'Restaurant_longitude'],
                               df.loc[i,'Delivery_location_latitude'],
                               df.loc[i,'Delivery_location_longitude'])
```
